```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv('/content/Automobile.csv')
```

```python
print(df.isnull().sum())
```

```python
numeric_cols = df.select_dtypes(include='number').columns.tolist()
```

```python
imputer = SimpleImputer(strategy='mean')
df[numeric_cols] = imputer.fit_transform(df[numeric_cols])
```

```python
X = df[['horsepower']]
y = df['mpg']
```

```python
model = LinearRegression()
```

```python
# Fit the model
model.fit(X, y)
```

```python
predicted_mpg = model.predict(X)
```

```python
plt.scatter(X, y, color='yellow', label='Actual data')
plt.plot(X, predicted_mpg, color='red', label='Linear Regression')
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.title('Linear Regression: MPG vs Horsepower')
plt.legend()
plt.show()
```
