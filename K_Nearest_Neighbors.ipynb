```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv("/content/Mall_Customers.csv")
```

```python
df.head()
```

```python
FEATURES = ['CustomerID','Age','Annual_Income_(k$)','Spending_Score']
X= df[FEATURES].copy()
Y= df['Genre']

knn= KNeighborsClassifier(n_neighbors=4)
```

```python
knn.fit(X,Y)
```

```python
new=pd.read_csv('/content/Mall_Customers.csv')
new
```

```python
new['pred_app']=knn.predict(new[FEATURES])
new
```
